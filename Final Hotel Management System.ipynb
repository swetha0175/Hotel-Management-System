```python
import pandas as pd
import os
```

```python
def loadHotelBooking():
    global HotelBookingDF
    HotelBookingDF = pd.read_excel("hotel_room_occupancy.xlsx",dtype=object)
```

```python
#checking for available room
# this shows the available room in the hotel that we can use to book the customers 
def checkAvRoom():
    AvailableRoom = HotelBookingDF[HotelBookingDF['Name'] == 'Not Occupied']
    print(AvailableRoom)
```

```python
# After using def checkAvRoom() to see which room are available we can use the below code to add the new -
# customer to the available room
# Allocating customer to room - we have not been told to add a customer to the available room 
# Therefore I am going to update and the change the Available default setting to the Customer Booking Information
def getNewCustomerBooking():
    Room_No = int(input("Enter the Room_No: "))
    if not HotelBookingDF['Room_No'].isin([Room_No]).any():
        print(f"The entered Room_No is non-existent.")
        return
    
    done = False
    while not done:
        try:
            CustomerName = input("Enter the name of the customer: ")
            Date = input("Enter the start booking date: ")
            Day =int(input("Enter the total days of stay: "))
            dateformat = pd.to_datetime(Date,format="%Y-%m-%d")
            done = True 
            
            HotelBookingDF.loc[HotelBookingDF['Room_No'] == Room_No, 'Name'] = CustomerName
            HotelBookingDF.loc[HotelBookingDF['Room_No'] == Room_No, 'Date_Occupancy'] = Date # Date Format = yyyy-mm-dd
            HotelBookingDF.loc[HotelBookingDF['Room_No'] == Room_No, 'Days_Occupancy'] = Day
            print("Room", Room_No, " has been added")
        except(ValueError):
            print("ValueError: Please Enter an INTEGER")
```

```python
#checking for occupied room
# this shows the occupied room by the existing customer 
def checkOccRoom():
    OccupiedRoom = HotelBookingDF[HotelBookingDF['Name'] != 'Not Occupied']
    print(OccupiedRoom)
```

```python
# we will be using Room_No to identify the updates needed for the people in that room 
# with the code below you can update the no.of day occupancy with the room number 
def UpdateDays():
    Room_No = int(input("Enter the Room_No: "))
    if not HotelBookingDF['Room_No'].isin([Room_No]).any(): # to check if the room number given is in the xlsx file 
        print(f"The entered Room_No is non-existent.")
        return
    
    UpdateChanges = "Days_Occupancy"
    if UpdateChanges == "Days_Occupancy":
        try:
            UpdatedDays=int(input("Updated Day Occupancy:"))
            HotelBookingDF.loc[HotelBookingDF["Room_No"] == Room_No, "Days_Occupancy"] = UpdatedDays
            print("The updated number of Day_Occupancy is:",UpdatedDays)
        except(ValueError): # to check if in case the person enters "Twenty" instead of "20"
            print("Please Enter an INTEGER.")
    else:
        pass
```

```python
# to delete the customer recorde and change it to the default value 
# I will be using the customer name to delete the data related
# Name = Not Occupied
# Date_ Occupancy = 01-01-22
# Days_Occupancy = 0 
def DefaultValue():
    DefaultName = "Not Occupied"
    DefaultDate = "2022-01-01"
    DefaultDays = 0
    dateformat = pd.to_datetime(DefaultDate,format="%Y-%m-%d") # to keep the format of the date 
    Room_No = int(input("Enter the Room_Number: "))
    if Room_No in list(HotelBookingDF['Room_No']):
        HotelBookingDF.loc[HotelBookingDF['Room_No'] == Room_No, 'Name'] = DefaultName
        HotelBookingDF.loc[HotelBookingDF['Room_No'] == Room_No, 'Date_Occupancy'] = DefaultDate
        HotelBookingDF.loc[HotelBookingDF['Room_No'] == Room_No, 'Days_Occupancy'] = DefaultDays
        print("Customer info has been deleted successfully.")
    else:
        print(f"The entered Room_No is non-existent.")
                                      
```

```python
def SaveUpdatedFile():
    filepath = Path('hotel_room_occupancy.xlsx')
    HotelBookingDF.to_excel(filepath)
    print("File has been saved at 'hotel_room_occupancy.xlsx'.")
```

```python
# display menu and get choice
def menu():
    print("\nMENU")
    print("A. Check for available rooms")
    print("B. Allot a room")
    print("C. Update occupancy for existing customer")
    print("D. Display occupied room")
    print("E. Checkout Customer")
    print("F. Save Updated file")
    print("Q. Quit application")
    choice = input("SELECT CHOICE: ")
    print("\n")
    return choice
```

```python
loadHotelBooking()

# despatch user's choice
selected = "X"
while selected != 'Q' and selected !='q':
    if selected == 'A' or selected =='a':
        checkAvRoom()
    elif selected == 'B' or selected =='b':
        getNewCustomerBooking()
    elif selected == 'C' or selected =='c':
        UpdateDays()
    elif selected == 'D' or selected =='d':
        checkOccRoom()
    elif selected == 'E' or selected =='e':
        DefaultValue()
    elif selected == 'F' or selected =='f':
        SaveUpdatedFile()
    else: pass

    selected = menu()
```
